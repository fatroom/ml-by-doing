```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define positive and negative classes

positive_points = np.array([(-2, 3), (0, 1), (2, -1)])
negative_points = np.array([(-2, 1), (0, -1), (2, -3)])
```

```python
# Plot initial data points
plt.figure(figsize=(8, 6))
plt.scatter(positive_points[:, 0], positive_points[:, 1], color='blue', marker='o', label='Positive class')
plt.scatter(negative_points[:, 0], negative_points[:, 1], color='red', marker='x', label='Negative class')
plt.legend(loc='best')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Initial Data Points')
plt.grid(True)
plt.show()
```

```python
# Combine data points and labels
X = np.vstack((positive_points, negative_points))
y = np.array([1] * len(positive_points) + [-1] * len(negative_points))
```

```python
def plot_decision_boundary(X, y, theta, theta0, iteration):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', marker='o', label='Positive class')
    plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', marker='x', label='Negative class')

    # Calculate decision boundary
    if theta[1] != 0:
        x_vals = np.linspace(-3, 3, 2)
        y_vals = -(theta[0] * x_vals + theta0) / theta[1]
        plt.plot(x_vals, y_vals, '--', color='green', label=f'Iteration {iteration}')
    else:
        plt.axvline(x=-theta0/theta[0], color='green', linestyle='--', label=f'Iteration {iteration}')

    plt.xlim(-3, 3)
    plt.ylim(-4, 4)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(f'Iteration {iteration}')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# Perceptron algorithm
def perceptron(X, y, steps=10):
    # Initialize weights and bias
    theta = np.zeros(X.shape[1])
    theta0 = 0

    for t in range(steps):
        for i, xi in enumerate(X):
            if y[i] * (np.dot(xi, theta) + theta0) <= 0:
                theta += y[i] * xi
                theta0 += y[i] 
                break 
        plot_decision_boundary(X, y, theta, theta0, t+1)
    return theta, theta0
```

```python
theta_final, thetha0_final = perceptron(X, y, steps=10)
```
